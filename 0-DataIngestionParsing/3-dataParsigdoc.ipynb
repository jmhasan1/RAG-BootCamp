### Word Document Processing### Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader , UnstructuredWordDocumentLoader
```

```python
# Method 1: Using Docx2txtLoader
print("Loading document using Docx2txtLoader...")
try:
    docx_loader = Docx2txtLoader("data/Air_Quality_Project_Report.docx")
    docs = docx_loader.load()
    print(f"Loaded {len(docs)} document(s) using Docx2txtLoader.")
    print(f"Content preview (first 500 characters):\n{docs[0].page_content[:500]}....")
    print(f"Metadata: {docs[0].metadata}")
except Exception as e:
    print(f"Error loading document with Docx2txtLoader: {e}")
```

```python
# Method 2 : Using UnstructuredWordDocumentLoader
print("\nLoading document using UnstructuredWordDocumentLoader...")
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/Air_Quality_Project_Report.docx" , mode="elements")
    # mode can be "elements" or "single" depending on how you want to structure the output documents (here we use "elements" to preserve the structure of the document)
    unstructured_docs = unstructured_loader.load()
    print(f"✅ Loaded {len(unstructured_docs)} elements")
    for i, doc in enumerate(unstructured_docs[:3]):
        print(f"\nElement {i+1}:")
        print(f"Type: {doc.metadata.get('category', 'unknown')}")
        print(f"Content: {doc.page_content[:100]}...")
except Exception as e:
    print(f"Error loading document with UnstructuredWordDocumentLoader: {e}")
```
